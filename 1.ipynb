```python
from agents.python_coder import generate_python_code
from agents.test_writer import generate_test_cases
from executor import run_tests
from agents.evaluator import generate_coverage_report_json

def main():
    # Step 1: Generate Python code for the given problem
    problem_description = "Write a Python function to check if string is a palindrome."
    generated_code = generate_python_code(problem_description)
    print("Generated Code:\n", generated_code)

    # Step 2: Generate test cases for the generated code
    test_cases = generate_test_cases(generated_code)
    print("Generated Test Cases:\n", test_cases)

    results = run_tests(generated_code, test_cases)
    print("Execution Results:\n", results)

    # Step 3: Evaluate the generated code using the test cases
    evaluation_result = generate_coverage_report_json(generated_code, test_cases)
    print("Evaluation Result:\n", evaluation_result)

if __name__ == "__main__":
    main()
```
